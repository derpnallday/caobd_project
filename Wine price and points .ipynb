```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
wine_reviews = pd.read_csv("winemag-data_first150k.csv")
wine_reviews.dropna() 
wine_reviews.drop(columns=['Unnamed: 0'])
del wine_reviews['Unnamed: 0']
wine_reviews.head()
```

```python
wine_reviews.describe()
```

### Distributions of Price and Points

```python
plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
plt.hist(wine_reviews['price'])
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Distribution of Wine Prices")

plt.subplot(1, 2, 2)
plt.hist(wine_reviews['points'])
plt.xlabel("Points")
plt.ylabel("Frequency")
plt.title("Distribution of Wine Review Points")

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(wine_reviews)
# shows the distribution for each individual variable as well as the relationship between the variables 
```

```python
sns.regplot(x='points', y='price', x_estimator=np.mean, data=wine_reviews)
plt.xlabel("Points")
plt.ylabel("Price")
plt.title("Relationship between Wine Price and Points")
plt.show()
# in general, as the price increases the points also increases, 
# but once it gets to a certain price the points cant get much better
```

```python
# prices of wines that received perfect points 
perfect_score =  wine_reviews['points']==100
perfect_wines = wine_reviews[perfect_score]
perfect_wines.sort_values('price', ascending=False)
```

```python
plt.hist(perfect_wines['price'])
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Wines with 100 Points")
plt.show()
```

### Average prices per country

```python
wine_reviews.groupby(['country']).mean().sort_values('price', ascending=False)
# US-France has the highest average price for wine
# England has highest average points but the price is not number 1 
```
